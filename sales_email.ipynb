```python
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.5,
    groq_api_key=groq
)
llm
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://eyglobal.yello.co/jobs/JUJGwtU0HNgGRkdebAT8BA?job_board_id=c1riT--B2O-KySgYWsZO1Q")
job_data = loader.load()
job_data
```

```python
!pip install langchain langchain-core
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()

prompt_template = PromptTemplate.from_template(
    "You are an expert at extracting data from webpage data. "
    "Extract the role, skills and job description in strictly JSON format "
    "from the data below. NO PREAMBLE.\n\n{job_description}"
)
chain = prompt_template | llm | parser
result = chain.invoke({
    "job_description":job_data
})
print(result)
```

```python
!pip install pandas 

import pandas as pd

df = pd.read_csv("project_based_skills.csv")
df
```

```python
text = df["TechSkills"].astype(str).tolist()
type(text)
text
```

```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = embedder.encode(text)
embeddings
```

```python
!pip install pinecone


```

```python
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone(api_key=pinecone_api_key)
index_name = "techskills-index"

if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,       
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )

index = pc.Index(index_name) 
index
```

```python
vectors = []

for i, row in df.iterrows():
    vectors.append({
        "id":str(i),
        "values": embeddings[i].tolist(),
        "metadata":{
            "techskills": row["TechSkills"],
            "link": row["Link"]
        }
    })

index.upsert(vectors)
```
